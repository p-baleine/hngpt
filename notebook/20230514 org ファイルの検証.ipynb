```python
from langchain.callbacks import get_openai_callback
from langchain.chat_models import ChatOpenAI
from tqdm.auto import tqdm

from hngpt.chain import ReviewerChain

llm = ChatOpenAI(temperature=0)
chain = ReviewerChain(llm=llm, verbose=False)
results = []

with get_openai_callback() as cb:
    for story in tqdm(topstories[:3]):
        review = chain.run(story=story)
        results.append((story, review))

print(f"Token usage: {cb.total_tokens} tokens.")
```

```python
for story, review in results:
    print(f"Title: {story.title}")
    print(f"URL: {story.url}")
    print(f"Review score: {review.score}")
    print(f"Review comment: {review.reason}")
    print()
```
